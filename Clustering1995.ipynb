```python
%matplotlib notebook
from Entities.DataReader import DataReader
from HierarchialClustering import HierarchialClustering
from scipy.cluster.hierarchy import dendrogram
```

```python
filename = "dataset.xlsx"
sheet = "sheet1"
```

DataReader class present inside the Entities folder does the mapping of features and generates DataRow objects

```python
dr = DataReader(filename)
dataRows = dr.read(sheet)
```

```python
#filter the claims from year 1995
X = [row for row in dataRows if row.policy.issue_date.year==1995]
Y = [row.id for row in X]
```

HierarchialClustering has an implementation of single-linkage clustering and uses the similarity measure defined by DataRow

```python
#perform clustering
H = HierarchialClustering()
Z = H.fit(X)
```

The output of the clustering algorithm is a linkage matrix which is fed to scipy dendogram for visualization

```python
d = dendrogram(Z, labels=Y,orientation='top')
```
